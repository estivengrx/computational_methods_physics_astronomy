```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

# Tarea
a) Realice la iterpolación de hermite usando scipy para los puntos de la clase antior

$x=[-1,0,1]$, $y=[2,1,2]$, $f'=[-8,0,8]$

b) Considere x=linspace(0,2*np.pi,10) y use

y=np.sin(x)

dy=np.cos(x)

para encontrar los puntos $x,y$ y $y'$ luego

Encuentre el polinomio de interpolación de hermite para dichos puntos y grafique el polinomio resultante.

```python
# a) Interpolación de Hermite datos pasados
x, y, dy = [-1, 0, 1], [2, 1, 2], [-8, 0, 8]
x_new = np.linspace(-1, 1, 100)

f_hermite1 = interpolate.CubicHermiteSpline(x, y, dy) # función de interpolación

plt.scatter(x, y, color='red', label='Datos')
plt.plot(x_new, f_hermite1(x_new), label='Interpolación de Hermite')

plt.title('Interpolación de datos con Hermite')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.grid(True, alpha=0.5)
plt.legend()
plt.show()
```

```python
# b) Interpolación Hermite otros datos
x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)
dy = np.cos(x)
x_new = np.linspace(-1, 7, 100)

f_hermite2 = interpolate.CubicHermiteSpline(x, y, dy) # función de interpolación

plt.scatter(x, y, color='red', label='Datos')
plt.plot(x_new, f_hermite2(x_new), label='Interpolación Hermite')

plt.title('Interpolación de datos con Hermite')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.grid(True, alpha=0.5)
plt.legend()
plt.show()
```
