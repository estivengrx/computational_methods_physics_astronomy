### **Regla de Simpson 3/8**

Es el resultado cuando para el integral se utiliza el resultado de una interpolaci√≥n con polinomio de tercer grado.

$$\int_a^bf(x)dx=\frac{3h}{8}[f(x_0)+3f(x_1)+3f(x_2)+f(x_3)] - \frac{3}{80}h^5f^{(4)}(z)$$

Donde $f^{(4)}(z) = \frac{d^4f}{dz^4}$.

## Tarea
Implemente la regla de Simpson 3/8 en Python

```python
def Simpson3_8(f, a, b, particiones):
   h = (b - a) / particiones
   S0 = f(a) + f(b)
   S1 = 0
   S2 = 0
   S3 = 0
   S4 = 0
   for i in range(1, particiones):  # suma en 1, ..., n-1
        if (i%2==0) and (i % 3 == 0):
            S1 += f(a + i*h) # suma de pares donde i es multiplo de tres
        elif (i%2 == 0) and (i % 3 != 0):
            S2 += f(a + i*h) # suma de pares donde i no es multiplo de tres
        elif (i%2 != 0) and (i % 3 != 0):
            S3 += f(a + i*h) # suma de impares donde i no es multiplo de tres
        elif (i%2 != 0) and (i % 3 == 0):
            S4 += f(a + i*h) # suma de impares donde i no es multiplo de tre

   return (S0 + S1 + 3*S2 + 3*S3 + S4)* (3*h / 8)
```

```python
f = lambda x: 3 * x + 2
f1 = lambda x: (x ** 4 * (1 - x) ** 4) / (1 + x**2)
```
