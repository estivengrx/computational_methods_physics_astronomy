### Tarea:

Implemente en Python funciones que le permite calcular la segunda derivada central, adelante y atrás, y haga un análisis similar para comparar cuál es más precisa.  Compare con la segunda derivada analítica del ejercicio de clase.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def e(x): #f(x)
    return np.exp(np.sin(x))

def de(x): #f'(x)
    return np.cos(x)*np.exp(np.sin(x))

def de2(x): #f''(x)
    return np.exp(np.sin(x)) * (np.cos(x)**2 - np.sin(x))
```

```python
def segunda_derivada(funcion, h=10e-9, punto:float=1, metodo: str='adelante'):
    '''
    Función para calcular la segunda derivada por tres métodos en un punto: 'adelante', 'atras' y 'central'
    '''

    if metodo == 'adelante':
        return (2 * funcion(punto) - 5 * funcion(punto + h) + 4 * funcion(punto + 2*h) - funcion(punto + 3*h)) / h**2
    elif metodo == 'atras':
        return (2 * funcion(punto) - 5 * funcion(punto - h) + 4 * funcion(punto - 2*h) - funcion(punto - 3*h)) / h**2
    elif metodo == 'central':
        return (funcion(punto + h) - 2 * funcion(punto) + funcion(punto - h)) / h**2
```

```python
a=-np.pi
b=np.pi
h=0.1 #tamaño del paso
N=round((b-a)/h) #redondear el resultado, el cual será la cantidad de datos
x = np.linspace(a,b,N)

# Definición de métodos y estilos de gráfica a utilizar
metodos = ['adelante', 'atras', 'central']
estilos = ['r+', 'g.', 'm*']

# Función con la segunda derivada en cada método
plt.figure(figsize=(15, 8))
plt.plot(x, de2(x), label='Segunda derivada')

for metodo, estilo in zip(metodos, estilos): # Cálculo de derivadas por métodos y graficas
    puntos_derivada = []
    for i in x:
        puntos_derivada.append(segunda_derivada(e, h, i, metodo))
    plt.plot(x, puntos_derivada, estilo, label=f'{metodo.capitalize()}', ms=4, linestyle='dashed')

plt.legend()
plt.grid(True, alpha=0.5)
plt.tight_layout()
plt.show()
```

En general, la segunda derivada central es la que más se aproxima al valor real de la gráfica analítica.
