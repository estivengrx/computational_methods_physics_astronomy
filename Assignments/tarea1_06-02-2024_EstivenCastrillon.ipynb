1.Escribir un programa que pregunte al usuario su edad y muestre por pantalla si es mayor de edad o no.

```python
edad_usuario = int(input("Ingrese su edad: "))
if (edad_usuario >= 18) and (edad_usuario in range(0, 110)):
    print("Usted es mayor de edad.")
else:
    print("Usted es menor de edad.")
```

2.Escribir un programa que pida al usuario un número entero y muestre por pantalla si es par o impar.

```python
numero_usuario = int(input("Ingrese un número entero: "))
if (numero_usuario % 2 == 0):
    print("El número que ingresó es par.")
else:
    print("El número que ingresó es impar.")
```
