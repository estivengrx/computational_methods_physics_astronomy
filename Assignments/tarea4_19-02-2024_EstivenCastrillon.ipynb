```python
dic_data = exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
colnames = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
data = pd.DataFrame(dic_data, index=colnames)
data
```


**4** Selecciones las filas donde score se encuentre entre 15 y 20 (incluidos) y
Cambie el score en la fila 'd' a 11.5


```python
# Seleccionando las filas donde score se encuentre entre 15 y 20 (incluidos)
data_score_filter = data.query(' score >= 15 and score <=20')
print('Scores between 15 and 20 included:')
print(data_score_filter, ' \n')

# Cambiando el score en la fila 'd' a 11.5
data_changed_score = data.copy()
data_changed_score.loc['d', 'score'] = 11.5
print('Data changed score d to 11.5:')
print(data_changed_score)
```

**5** Agregue una nueva fila con etiqueta 'k' al data frame con un valor que usted desee para cada columna luego Borre la fila 'k' y entregue nuevamente el DataFrame original.

```python
# Añadiendo nueva fila
nueva_fila = {'name': 'Estiven', 'score': 15.5, 'attempts': 2, 'qualify': 'yes'}
data_nueva_fila = data.copy()
data_nueva_fila.loc['k'] = nueva_fila
print('Dataframe con nueva fila "k"')
print(data_nueva_fila, '\n')

# Borrando la fila 'k'
print('Eliminando la fila "k" del dataframe')
print('Entregando el dataframe original:')
data_nueva_fila.drop('k', axis=0, inplace=True)
print(data_nueva_fila)
```

**6**Ordene el DataFrame por nombre ('name') en orden ascendente y reemplace los valores de la columna 'qualify' que contiene los valores 'yes' y 'no' por los valores buleanos True and False respectivamente.

```python
data_sorted_and_boolean = data.copy()
data_sorted_and_boolean.sort_values('name', ascending=True)
print('Dataframe ordenado en ascendente:')
print(data_sorted_and_boolean, '\n')

data_sorted_and_boolean['qualify'] = (data_sorted_and_boolean['qualify'] == 'yes')
print('Dataframe con la columna qualify en boolean:')
print(data_sorted_and_boolean)
```
