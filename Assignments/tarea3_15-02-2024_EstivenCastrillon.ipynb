```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

lista = []  #lista de valores finales

# Realizando 1000 caminatas aleatorias
for _ in range(1000):

    posicion = 0 #reiniciando la posición de inicio para cada caminata

    # Realizando 100 pasos en cada caminata aleatoria
    for _ in range(100):
        # Generando un paso aleatorio (-1 o 1) y actualizando la posición
        valor_aleatorio = rd.choice([-1, 1])
        posicion += valor_aleatorio

    # Agregando la posición final después de la caminata a la lista
    lista.append(posicion)

lista = np.array(lista)

sns.histplot(lista, bins=20, color='blue', edgecolor='black', kde=True)
plt.xlabel('Posición final')
plt.ylabel('Frecuencia')
plt.title('Simulación borracho')
plt.show()
```
